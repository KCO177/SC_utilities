```python
from youtube_transcript_api import YouTubeTranscriptApi
from youtube_transcript_api._errors import NoTranscriptFound, TranscriptsDisabled, VideoUnavailable

def get_czech_transcript(video_id):
    try:
        ytt_api = YouTubeTranscriptApi()
        transcript = ytt_api.fetch(video_id, languages=['cs'])  # Fetch Czech transcript if available
        print("📜 Český přepis nalezen:")
    except NoTranscriptFound:
        print("❌ Český přepis nebyl nalezen.")
        return
    except TranscriptsDisabled:
        print("❌ Přepisy jsou zakázány pro toto video.")
        return
    except VideoUnavailable:
        print("❌ Video není dostupné nebo neexistuje.")
        return

    # Print only the transcript text, no time info
    for entry in transcript:
        print(entry.text)

# Example video ID (replace with a Czech-language video ID)
video_id = "Hr79FiK_UU4"
get_czech_transcript(video_id)
```
